```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
Games = pd.read_csv(r"C:\Users\nicol\Documents\Twitch Project\Twitch_game_data.csv", encoding = 'cp1252')
# credit: https://www.kaggle.com/datasets/rankirsh/evolution-of-top-games-on-twitch?resource=download
```

```python
avg_peak = sns.scatterplot(data=Games, x="Avg_viewers", y="Peak_viewers", hue="Avg_viewers").set(title ='Average Viewers vs. Peak Viewers', xlabel='Average Viewers', ylabel='Peak Viewers')
```

```python
sns.set_theme(style="whitegrid")

cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
streamer_peak = sns.relplot(
    data=Games,
    x="Streamers", y="Peak_viewers",
    hue="Avg_viewers", size="Avg_viewers",
    palette=cmap, sizes=(10, 200),
)
streamer_peak._legend.set_title("Avg. Viewers")
streamer_peak.set(xscale="log", yscale="log")
streamer_peak.ax.xaxis.grid(True, "minor", linewidth=.25)
streamer_peak.ax.yaxis.grid(True, "minor", linewidth=.25)
streamer_peak.despine(left=True, bottom=True)
streamer_peak.set(ylabel = "Peak Viewers", title = "Streamers vs. Peak Viewers")
```

```python
plt.figure(figsize=(10,8))
top_games = sns.barplot(x = 'Game',y = 'Streamers',data = Games.head(20))
plt.xticks(rotation = 90)
top_games.set(title = "Top 20 Games and Number of Streamers")
```
